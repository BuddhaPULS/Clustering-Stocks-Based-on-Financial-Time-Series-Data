```python
import pandas as pd
from tqdm import tqdm 
import akshare as ak
import time
```

```python
stocks = [
    '000063','000333','000617','000725','000792','000938',
    '002027','002230','002371','002475','002594','002714',
    '300015','300124','300408','300750','300760',
    '600009','600019','600028','600030','600031','600036',
    '600048','600111','600276','600309','600406','600415',
    '600426','600436','600519','600585','600660','600887',
    '600893','600900','600941','601012','601088','601318',
    '601600','601668','601766','601816','601888','601899',
    '603259','605499','688981'
]
```

```python
all_fin=[]
```

```python
for s in tqdm(stocks, desc="Fetching financial data"):
    try:
        time.sleep(1.5)
        df_f = ak.stock_financial_abstract(symbol=s)
        if df_f is None or df_f.empty:
            continue
        df_f = df_f.drop(columns=['选项'], errors='ignore')
        df_f = df_f.set_index('指标')
        df_f=df_f.drop_duplicates()
        df_f=df_f.transpose()
        df_f.index = pd.to_datetime(df_f.index)
        df_f['symbol'] = s
        df_f = df_f.reset_index().rename(columns={'index':'report_date'})
        all_fin.append(df_f)
    except Exception as e:
        print(f"Skip {s}: {e}")
    
```

```python
df=pd.concat(all_fin, ignore_index=True)
```
