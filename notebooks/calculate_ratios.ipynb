```python
import pandas as pd
```

```python
financial_data=pd.read_parquet('financial_data.parquet')
df=financial_data.copy()
```

```python
used_features=['symbol',
               'report_date',
               '资产负债率',
               '经营活动净现金/销售收入',
               '营业总收入',
               '扣非净利润',
               '每股未分配利润',
               '净利润',
               '股东权益合计(净资产)',
               '每股净资产',
               '归母净利润',
               '每股现金流量净额']
```

```python
df_subset=df[used_features]
```

```python
print("Dataframe info:")
df_subset.info()
print('\n Dataframe description:')
df_subset.describe()
```

```python
df['资产负债率'].isnull().sum()
```

```python
#1.DE：total liablity/net worth
df['DE']=df['debt_ratio']*0.01/(1 - df['debt_ratio']*0.01)
```

```python
#2.NCOS:net cash from operation/sales
df['NCOS']=df['经营活动净现金/销售收入']
```

```python
#3.pretax net Non-operating income
df["PNNIS"]=df['扣非净利润']/df['营业总收入']
```

```python
#4. (net income -divid)/net income

df['NDN']=df['每股未分配利润']/(df['净利润']/(df["股东权益合计(净资产)"]/df['每股净资产']))
```

```python
# 5. sales/total assets
df['STA']=df['营业总收入']/df['股东权益合计(净资产)']
```

```python
#6.ICSNW: net income -dividends/shareholder equity
df['ICSNW']=df['扣非净利润']/df['股东权益合计(净资产)']
```

```python
# 7.NITN: real net income to net income
df['NITN']=df['归母净利润']/df['净利润']
```

```python
#8.Cash per share
df['CASH_PER_SHARE']=df['每股现金流量净额']
```

```python
output_features=['DE',
                 'NCOS',
                 'PNNIS',
                 'NDN',
                 'STA',
                 'ICSNW',
                 'NITN',
                 'CASH_PER_SHARE']
```

```python
# rename used features
rename_dict = {
    '资产负债率': 'debt_ratio',
    '经营活动净现金/销售收入': 'op_cash_to_sales',
    '营业总收入': 'operating_revenue',
    '扣非净利润': 'net_profit_excl_extraordinary',
    '每股未分配利润': 'eps_retained_earnings',
    '净利润': 'net_profit',
    '股东权益合计(净资产)': 'total_equity',
    '每股净资产': 'eps_equity',
    '归母净利润': 'parent_net_profit',
    '每股现金流量净额': 'eps_cash_flow'
}
df = df.rename(columns=rename_dict)
```

```python
# after rename, basic features 
renamed_features = [
    'symbol',
    'report_date',
    'debt_ratio',
    'op_cash_to_sales',
    'operating_revenue',
    'net_profit_excl_extraordinary',
    'eps_retained_earnings',
    'net_profit',
    'total_equity',
    'eps_equity',
    'parent_net_profit',
    'eps_cash_flow'
]
```

```python
features=renamed_features+output_features
```

```python
# contain all used features and ratios
output_df=df[features]
```

```python
print('\n Output dataframe info:')
output_df.info()
print('\n output dataframe description')
output_df.describe()
```

```python
output_df.to_parquet('financial_ratios.parquet', index=False)
print("\n used financial data and ratios data saved to 'financial_ratios.parquet'")
```

```python
ratio_df=df[['symbol','report_date']+output_features]
```

```python
ratio_df.info()
```

```python
ratio_df.to_parquet('ratio_data.parquet', index=False)
print("\n ratio data saved to 'ratio_data.parquet'")
```
